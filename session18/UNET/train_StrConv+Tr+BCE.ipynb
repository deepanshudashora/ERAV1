# Importing Libaries

```python
import warnings
warnings.filterwarnings("ignore")

import os 
import torch

import lightning.pytorch as pl
from torchood.data.oxfordpets_module import OxfordIiitPetsDataModule
from torchood.models.unet_module import UNetLitModule
```

```python
config = {
    "project_name" : "Unet-Oxford-Iiit-Pets", # Wandb project name
    "run_name" : "StrConv+tr+ce", # wandb run name
    "num_epochs" : 50,
    "lr": 1e-3
}
```

```python
datamodule = OxfordIiitPetsDataModule(
    image_dir="datasets/oxford_iiit_pets/images", 
    mask_dir="datasets/oxford_iiit_pets/annotations/trimaps",
    batch_size=16,
    num_workers=8,
    pin_memory=True
)
```

```python
model = UNetLitModule(learning_rate=config["lr"], out_channels=3, strided_conv=True, criterion_type="ce")
```

```python
from lightning.pytorch.callbacks import LearningRateMonitor
from lightning.pytorch.loggers import WandbLogger


callbacks = [LearningRateMonitor(logging_interval="step")]
logger = WandbLogger(project=config["project_name"], name=config["run_name"])
trainer = pl.Trainer(
    callbacks=callbacks,
    logger=logger,
    max_epochs=config["num_epochs"],
    accelerator = "gpu",
)
trainer.fit(model=model, datamodule=datamodule)
```

```python
import matplotlib.pyplot as plt

model.to("cuda")
for i, (x, y) in enumerate(datamodule.val_dataloader()):
    # taking only first image from batch
    x = x[0].unsqueeze(0)
    y = y[0]

    pred = model(x.to("cuda"))
    pred = pred.argmax(dim=1).squeeze().cpu().numpy()

    plt.subplot(1, 3, 1)
    plt.imshow(x[0].permute(1, 2, 0))
    plt.axis("off")
    
    plt.subplot(1, 3, 2)
    plt.imshow(y)
    plt.axis("off")

    plt.subplot(1, 3, 3)
    plt.imshow(pred)
    plt.axis("off")

    plt.show()
    if i==10:
        break
```
